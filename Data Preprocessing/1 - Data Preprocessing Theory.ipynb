## What is Numerical Variable?
A numerical variable is a variable where the measurement or number has a numerical meaning.

### What is Discret Variable?
The value which can be counted as whole value (fixed) called as discret variable. Ex 1, 45, 73, 23 not 23.4, 56.76, 3.04, (integer)

The value contain range, measurements called as continuous variable. Ex 1.3, 4.5, 73.67, 20.3 not 23, 56, 3 (Floating Value)

## What is Categorical Variable?
The values of a categorical variable are selected from a group of categories also called labels.

### What is Ordinal Variables?
Ordinal variables are categorical variables in which the categories can be meaningfully ordered. When the ordinal variable convert into the number then it has mathematical value. Ex. Rating of product.

## What is missing value/data?
In the variable or any observation, values are not stored called as missing values/data.

### What are the methods to handle missing values?
- Ignore missing values row / Delete row
- Fill missing value manually
- Global constant
- Measure of central tendency (Mean, Midian & Mode)
- Measure of central tendency for each class
- Most probable value (ML Algorithms - Linear Regg, Decision Tree)
