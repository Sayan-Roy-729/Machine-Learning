## What are the methods to handle missing values?
- Ignore missing values row / delete row
- Fill missing value manually
- Global constant
- Measure of central tendency (Mean, Midian & Mode)
- Measure of central tendency for each class
- Most probable value (ML Algorithms)

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# load dataset
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv("../DataSets/house-prices-advanced-regression-techniques/train.csv")
df.head()
```

```python
# info of the dataset
df.info()
```

```python
# check null values
df.isnull().sum()
```

```python
# Which variable has x% missing values
missing_value_per = df.isnull().sum()/df.shape[0] * 100
missing_value_per
```

```python
# Which columns have more that 20% missing values and then drop these
missing_value_clm_gre_20 = missing_value_per[missing_value_per > 20].keys()
missing_value_clm_gre_20
```

```python
df2_drop_clm = df.drop(columns=missing_value_clm_gre_20)
df2_drop_clm.head()
```

```python
# Extract the numeric columns
df3_num = df2_drop_clm.select_dtypes(include=['int64', 'float64'])
df3_num.head()
```

```python
# Check is there is any missing values are present in out numeric data
plt.figure(figsize=(16, 9))
sns.heatmap(df3_num.isnull())
plt.show()
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum() > 0]
missing_num_var
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for index, variable in enumerate(missing_num_var):
    plt.subplot(2, 2, index+1)
    sns.distplot(df3_num[variable], bins=20, kde_kws={'linewidth': 5, 'color': '#DC143C'})

plt.show()
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth': 8, 'color': 'red'}, label='Original')
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth': 5, 'color': 'green'}, label="Mean")
    plt.legend()
plt.show()
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df3_num[var], bins=20, hist=False, kde_kws={'linewidth': 8, 'color': 'red'}, label='Original')
    sns.distplot(df4_num_mean[var], bins=20, hist=False, kde_kws={'linewidth': 5, 'color': 'green'}, label="Mean")
    sns.distplot(df5_num_median[var], bins=20, hist=False, kde_kws={'linewidth': 5, 'color': 'k'}, label="Median")
    plt.legend()
plt.show()
```

```python
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10, 10))
    plt.subplot(3, 1, 1)
    sns.boxplot(df[var])
    plt.subplot(3, 1, 2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3, 1, 3)
    sns.boxplot(df5_num_median[var])

plt.show()
```

```python
df_concat = pd.concat([df3_num[missing_num_var], df4_num_mean[missing_num_var], df5_num_median[missing_num_var]], axis=1)
df_concat[df_concat.isnull().any(axis=1)]
```
