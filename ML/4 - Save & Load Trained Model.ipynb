```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../DataSets/homeprices.csv")
df.head()
```

```python
model = LinearRegression()
model.fit(df[['area']], df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

```python
# Save model using pickle module
import pickle
with open("model_pickle.pkl", "wb") as f:
    pickle.dump(model, f)

# Load the model
with open("model_pickle.pkl", "rb") as f:
    mp = pickle.load(f)
    
mp.predict([[5000]])
```

```python
# save model using sklearn joblib
from sklearn.externals import joblib
joblib.dump(model, 'model_joblib') # store as a binary file

# load the model
mj = joblib.load("model_joblib")
ml.predict([[5000]])
```
