```python
import pandas as pd
```

```python
df = pd.read_csv("../DataSets/salaries.csv")
df.head()
```

```python
inputs = df.drop("salary_more_then_100k", axis=1)
target = df["salary_more_then_100k"]

# encode the "company" column
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()

inputs["company_n"] = le_company.fit_transform(inputs["company"])
inputs["job_n"] = le_job.fit_transform(inputs["job"])
inputs["degree_n"] = le_degree.fit_transform(inputs["degree"])

inputs.head()
```

```python
inputs_n = inputs.drop(columns=["company", "job", "degree"], axis=1)
inputs_n.head()
```

```python
# create decision tree classification model
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

# train the model
model.fit(inputs_n, target)

# predict the score
model.score(inputs_n, target)
```

```python
# predict
model.predict([[2, 2, 1]])  # company=google, job=sales executive degree=masters
```
