# Matplotlib
## What is matplotlib?
- Matplotlib is 2D and 3D plotting python library.
- It produce high quality graph.
- It also supports animation.

## Graphs
- Histogram
- Bar Charts
- Scatterplot
- Area Plot
- Pie plot
- Error charts
- Power Spectra

## Line Plot

```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
days = [i for i in range(1, 16)]  # Create 1 to 15 days
temperature = [36.6, 37, 37.7, 39, 40.1, 43, 43.4, 45, 45.6, 40.1, 44, 45, 46.8, 47, 47.8]

plt.plot(days, temperature)
plt.title('Delhi Temperature')
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
days = [i for i in range(1, 16)]  # Create 1 to 15 days
temperature = [36.6, 37, 37.7, 39, 40.1, 43, 43.4, 45, 45.6, 40.1, 44, 45, 46.8, 47, 47.8]

plt.plot(days, temperature)
plt.axis([0, 30, 0, 50]) # Change the range of axis values axis(xmin, xmax, ymin, ymax)
plt.title('Delhi Temperature')
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
days = [i for i in range(1, 16)]  # Create 1 to 15 days
temperature = [36.6, 37, 37.7, 39, 40.1, 43, 43.4, 45, 45.6, 40.1, 44, 45, 46.8, 47, 47.8]

plt.plot(days, temperature, color="g", marker='o', linestyle="--", linewidth=2, markersize=10)
plt.title('Delhi Temperature')
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
days = [i for i in range(1, 16)]  # Create 1 to 15 days
temperature = [36.6, 37, 37.7, 39, 40.1, 43, 43.4, 45, 45.6, 40.1, 44, 45, 46.8, 47, 47.8]

plt.plot(days, temperature, "go--", linewidth=2, markersize=10, label="Temp line")
plt.title('Delhi Temperature', fontsize=15)
plt.xlabel("days", fontsize=15)
plt.ylabel("temperature", fontsize=15)
plt.legend(loc=4)
plt.show()
```

```python
days = [i for i in range(1, 16)]  # Create 1 to 15 days
temperature = [36.6, 37, 37.7, 39, 40.1, 43, 43.4, 45, 45.6, 40.1, 44, 45, 46.8, 47, 47.8]

style.use("ggplot")
plt.plot(days, temperature, "mo--", linewidth=2, markersize=10, label="Temp line")
plt.title('Delhi Temperature', fontsize=15)
plt.xlabel("days", fontsize=15)
plt.ylabel("temperature", fontsize=15)
plt.legend(loc=4)
plt.grid(color='k', linestyle="-", linewidth=2)
plt.show()
```

```python
days = [i for i in range(1, 16)]  # Create 1 to 15 days
delhi_temp = [36.6, 37, 37.7, 39, 40.1, 43, 43.4, 45, 45.6, 40.1, 44, 45, 46.8, 47, 47.8]
mumbai_temp = [39, 39.4, 40, 40.7, 41, 42.5, 43.5, 44, 44.9, 44, 45, 45.1, 46, 47, 46]

style.use("ggplot")
plt.plot(days, delhi_temp, "mo--", linewidth=2, markersize=10, label="Delhi")
plt.plot(days, mumbai_temp, "yo:", linewidth=2, markersize=10, label="Mumbai")
plt.title('Delhi vs Mumbai Temperature', fontsize=15)
plt.xlabel("days", fontsize=15)
plt.ylabel("temperature", fontsize=15)
plt.legend(loc=4)
plt.grid(color='w', linestyle="-", linewidth=2)
plt.show()
```
