```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]

plt.figure(figsize=(16, 9))
explode = [0.03, 0, 0.1, 0, 0]
colors = ["c", "b", "r", "y", "g"]
textprops = {"fontsize": 15}
plt.pie(class1_students, labels=classes, explode=explode, colors=colors, autopct="%0.2f%%", shadow=True, 
        radius=1.4, startangle=90, textprops=textprops)
plt.show()
```

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]

plt.figure(figsize=(16, 9))
explode = [0.03, 0, 0.1, 0, 0]
colors = ["c", "b", "r", "y", "g"]
textprops = {"fontsize": 15}
wedgeprops = {"linewidth": 4, "width": 1, "edgecolor": "k"}
plt.pie(class1_students, labels=classes, explode=explode, colors=colors, autopct="%0.2f%%", shadow=True, 
        radius=1.4, startangle=90, textprops=textprops, pctdistance=0.6, labeldistance=1.4, counterclock=True,
       wedgeprops=wedgeprops, center=(2, 3), frame=True, rotatelabels=True)
plt.show()
```
