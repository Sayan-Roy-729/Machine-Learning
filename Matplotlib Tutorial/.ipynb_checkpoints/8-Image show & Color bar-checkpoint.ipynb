```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas
```

```python
img = mpimg.imread("pie_chart.png")
print("Type: ", type(img))
print("Shape: ", img.shape)
print("Dimension: ", img.ndim)
```

```python
plt.figure(figsize=(16, 9))
plt.imshow(img)
plt.axis("off")  # Disable the axis show
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.imshow(img)
plt.axis("off")  # Disable the axis show
plt.colorbar() # Enable the color bar
plt.show()
```

```python
single_channel = img[:, :, 1]

plt.figure(figsize=(16, 9))
# camp or color map only available for two dimensional values
plt.imshow(single_channel, cmap="hot")
plt.axis("off")  # Disable the axis show
plt.colorbar() # Enable the color bar
plt.show()
```
