# Histogram

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18, 45, (100))
py_student_age = np.random.randint(15, 40, (100))
print("ml_student_age\n", ml_student_age)
print("py_student_age\n", py_student_age)
```

```python
plt.hist(ml_student_age, bins=10)
plt.title("ML Students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students")
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist(ml_student_age, bins=bins, rwidth=0.8, histtype='step', align='right', 
         orientation='horizontal', color='m', label='ML Student'
        )
plt.hist(py_student_age, bins=bins, rwidth=0.8, histtype='step', align='right', 
         orientation='horizontal', color='g', label='Py Student'
        )
plt.title("ML Students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students")
plt.legend()
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist(ml_student_age, bins=bins, rwidth=0.8, histtype='bar', align='right', 
         orientation='vertical', color='m', label='ML Student'
        )
plt.hist(py_student_age, bins=bins, rwidth=0.8, histtype='bar', align='right', 
         orientation='vertical', color='g', label='Py Student'
        )
plt.title("ML Students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students")
plt.legend()
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.figure(figsize=(16, 9))
plt.hist([ml_student_age, py_student_age], bins=bins, rwidth=0.8, histtype='bar', align='right', 
         orientation='vertical', color=['m', 'y'], label=['ML Student', 'Py Students'])
plt.title("ML Students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students")
plt.legend()
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

style.use('ggplot')
plt.figure(figsize=(16, 9))
plt.hist([ml_student_age, py_student_age], bins=bins, rwidth=0.8, histtype='bar', align='right', 
         orientation='vertical', color=['m', 'y'], label=['ML Student', 'Py Students'])
plt.title("ML Students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students")
plt.legend()
plt.show()
```
