# Bar Plot

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
# Dataset of 'Indian Artificial Intelligence Production (IAIP) Class'

# classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]

classes = ['Python', 'R', 'AI', 'ML', 'DS']
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students)
plt.show()
```

```python
plt.barh(classes, class1_students)
plt.show()
```

```python
plt.bar(classes, class1_students, width=0.2, align='edge', color='y', edgecolor='m', 
        linewidth=5, alpha=0.5, linestyle='--', label="Class 1 Students", visible=True)
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16, 9))
plt.bar(classes, class1_students, width=0.6, align='edge', color='k', edgecolor='m', 
        linewidth=5, alpha=0.5, linestyle='--', label="Class 1 Students", visible=True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.barh(classes, class1_students, align='edge', color='k', edgecolor='m', 
        linewidth=5, alpha=0.5, linestyle='--', label="Class 1 Students", visible=True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.bar(classes, class1_students, width=0.6, align='edge', color='k', edgecolor='m', 
        linewidth=5, alpha=0.5, linestyle='--', label="Class 1 Students", visible=True)
plt.title("Bar Chart of IAIP Class", fontsize=20)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No of Students", fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.bar(classes, class1_students, width=0.2, color='b', label="Class 1 Students")
plt.bar(classes, class2_students, width=0.2, color='g', label="Class 2 Students",)
plt.title("Bar Chart of IAIP Class", fontsize=20)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No of Students", fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
classes_index = np.arange(len(classes))
width = 0.2

plt.bar(classes_index, class1_students, width=width, color='b', label="Class 1 Students")

plt.bar(classes_index + width, class2_students, width=width, color='g', label="Class 2 Students")

plt.bar(classes_index + width + width, class3_students, width=width, color='y', label="Class 3 Students")

plt.xticks(classes_index + width, classes, rotation=20)  # Replace old label with our custom label
plt.title("Bar Chart of IAIP Class", fontsize=20)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No of Students", fontsize=15)
plt.legend()
plt.show()
```
