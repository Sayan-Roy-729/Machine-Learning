```python
df = pd.read_csv("C:\\Users\sayan\Documents\Machine Learning\DataSets\Google Play Store\googleplaystore.csv", nrows=1000)
df.shape
```

```python
df.head()
```

```python
x = df["Rating"]
y = df["Reviews"]
```

```python
plt.scatter(x, y)
plt.title("Google Play STore Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

From the plot, we can see that from 4.0 to near about 4.7 ratings, maximum persons done the reviews.

```python
plt.figure(figsize=(16, 9))
plt.scatter(x, y, c="r", marker="*", s=100, alpha=0.5, linewidths=10, edgecolors='g')
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.scatter(x, y, c="r", marker="*", s=100, alpha=0.5, linewidths=10, edgecolors='g')
plt.scatter(x, df["Installs"], c="y", marker="o", s=100, alpha=0.5, linewidths=10, edgecolors='c')
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews & Installs")
plt.show()
```
