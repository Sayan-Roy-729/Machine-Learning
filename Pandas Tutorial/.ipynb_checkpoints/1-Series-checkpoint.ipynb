# Series:

```python
import pandas as pd
```

```python
pd.__version__
```

## With Python List:

```python
list_s = [1, 2, -3, 6.2, 'data values']
series1 = pd.Series(list_s)
series1
```

```python
type(series1)
```

```python
series2 = pd.Series([1, 2, 3, 4])
series2
```

## Empty Series:

```python
empty_s = pd.Series([])
empty_s
```

## *index* parameter:

```python
series3 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
series3
```

## *dtype* parameter:

```python
series3 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'], dtype=float)
series3
```

## *name* parameter:

```python
series3 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'], dtype=float, name='data values')
series3
```

## Create Scalar Series: 

```python
scalar_s = pd.Series(0.5)
scalar_s
```

```python
scalar_s = pd.Series(0.5, index=[1, 2, 3])
scalar_s
```

## Create Series Using Python Dictionary:

```python
dict_s = pd.Series({'a': 1, 'b': 2, 'c': 3})
dict_s
```

```python
s4 = pd.Series([1, 2, 3, 4, 5])
s4
```

## Slicing:

```python
s4[0]
```

```python
s4[3]
```

```python
s4[0:3]
```

## min & max

```python
max(s4)
```

```python
min(s4)
```

## Arithmetic Operations 

```python
s4[s4 > 3]
```

```python
s5 = pd.Series([1, 2, 3, 4, 5])
s5
```

```python
s4 + s5
```

```python
s6 = pd.Series([1, 2, 3])
s6
```

```python
s5 + s6
```
