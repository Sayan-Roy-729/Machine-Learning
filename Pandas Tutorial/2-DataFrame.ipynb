# Pandas DataFrame
Pandas DataFrame is two-dimensional, size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows & columns).

```python
import pandas as pd
```

## Empty DataFrame

```python
empty_df = pd.DataFrame()
empty_df
```

## Create DataFrame from Python List

```python
lst = ['a', 'b', 'c']
df1 = pd.DataFrame(lst)
df1
```

```python
ls_of_ls = [[1, 2, 3], [2, 3, 4], [4, 5, 6]]
df2 = pd.DataFrame(ls_of_ls)
df2
```

## Create DataFrame from Python Dictionary

```python
dict1 = {"ID": [11, 22, 33, 44]}
df3 = pd.DataFrame(dict1)
df3
```

```python
dict1 = {"ID": [11, 22, 33, 44], "SN": [1, 2, 3, 4]}
df3 = pd.DataFrame(dict1)
df3
```

```python
ls_dict = [{'a': 1, 'b': 2}, {'a': 3, 'b': 4}]
df4 = pd.DataFrame(ls_dict)
df4
```

```python
ls_dict = [{'a': 1, 'b': 2}, {'a': 3, 'b': 4, 'c': 5}]
df4 = pd.DataFrame(ls_dict)
df4
```

## Create DataFrame Using Pandas Series

```python
dict_sr = {'Id': pd.Series([1, 2, 3]), 'SN': [111, 222, 333]}
df5 = pd.DataFrame(dict_sr)
df5
```
