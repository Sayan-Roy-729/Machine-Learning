```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
sns.distplot(tips_df['tip'])
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], bins=10, hist=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], bins=10, kde=False)
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], bins=10, rug=True)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.distplot(tips_df['total_bill'], bins=10, fit=norm, color="r", vertical=True, norm_hist=True, axlabel="Total Bill", 
             label="Total Bill")
plt.title("Histogram of Total Bill", fontsize=20)
plt.legend()
plt.show()
```
